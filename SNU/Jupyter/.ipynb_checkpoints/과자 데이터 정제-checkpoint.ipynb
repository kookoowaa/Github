```python
import json
```

```python
snack_txt = open('D:/snack.txt','r')
```

```python
snack_list = []

for i in snack_txt:
    try:
        snack_list.append(json.loads(i))
    except:
        continue
```

```python
snack_text = []
for i in snack_list:
    snack_text.append(i['text'])
```

```python
snack_text
```

```python
len(snack_text)
```

```python
snack_filtered = []
for i in snack_text:
    if '캠' not in i:
        if '추첨' not in i:
            snack_filtered.append(i.replace('\n',''))
```

```python
snack_filtered
```

```python
snack_final = list(set(snack_filtered))
```

```python
snack_final
```

```python
a = open('d:/snack_analysis.txt', 'a', encoding = 'UTF-8')
for i in snack_final:
    a.write(i)
    a.write('\n')
a.close()
```

```python
len(snack_final)
```

```python
snack_final = [i.replace('RT','') for i in snack_final]
```

```python
snack_final
```

```python
import re
```

```python
snack_final2 = []
for i in snack_final:
    snack_final2.append(re.sub('https[:a-zA-Z0-9./]*','',re.sub('@[a-zA-Z0-9_:…]*', '', i)))
```
