## 1. HDF5 형식 사용하기
- HDF5 (Hierarchical Data Format): 자바, 파이썬, MATLAB 언어를 위한 인터페이스를 제공하는 C 라이브러리
- PyTables와 h5py 두가지 라이브러리 사용 가능
- pandas는 PyTables를 이용한 HDFStore 라는 가벼운 사전 클래스를 통해 객체를 저장

```python
pd. HDFStore('mydata.h5')
```

```python
store = pd.HDFStore('mydata.h5')
```

```python
store['obj1'] = frame
```

```python
store['obj1_col'] = frame['a']
```

```python
store
```

```python
store['obj1']
```

- HDF5 파일에 포함된 객체는 파이썬 사전과 유사한 형식으로 사용 가능
- BA는 CPU 보다는 IO 성능에 의존적이므로 HDF5 같은 도구를 사용 시 성능 향상 가능
