# 4-7 계단 오르내리기 예제

```python
import numpy as np
import random
```

```python
position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
```

이하 예제

```python
nsteps = 1000
```

```python
draws = np.random.randint(0,2, size = nsteps)
```

```python
steps = np.where(draws>0, 1, -1)
```

```python
walk = steps.cumsum()
```

```python
walk.min() 
```

```python
walk.max()
```

```python
(np.abs(walk) >=10).argmax()
```

```python
np.arange(1,1001)*(np.abs(walk) >=10)
```

위의 시뮬레이션을 5000번 반복

```python
nwalks = 5000
nstep = 1000
draws = np.random.randint(0,2, size = (nwalks, nstep))
steps = np.where(draws==0, -1, draws)
walks = steps.cumsum(axis = 1)
walks
```

```python
walks.max()
```

```python
walks.min()
```

```python
hits30 = (np.abs(walks) >=30).any(1)
hits30
```

```python
sum(hits30)
```

```python
crossing_times = (np.abs(walks) >=30).argmax(axis = 1)
crossing_times
```

```python
crossing_times.mean()
```
