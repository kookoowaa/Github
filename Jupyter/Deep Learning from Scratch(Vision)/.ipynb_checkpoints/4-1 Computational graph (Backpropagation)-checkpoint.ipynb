# *오차역전파법*

## 1. 계산그래프

> - 오차역전파법을 이해하는 방법은 1)수식을 통한 방법과, **2)계산그래프**를 통한 방법이 있음
- 일반적으로 수식 중심으로 설명을 하지만, 시각적으로 계산 그래프를 사용해서 이해하는 방식이 더 쉽고 직관적임
- 계산그래프(computational graph)는 계산 과정을 그래프로 나타낸 것
- 복수의 노드(node)와 엣지(edge)로 시각화

___

### 1) 계산 그래프로 풀다

- 계산 그래프에 대한 이해를 돕기 위해 아래 예제들을 계산 그래프 표현
> *문제1: 현빈군은 슈퍼에서 1개에 100원인 사과를 2개 샀습니다. 이 때 지불금액을 구하세요. 단, 소비세가 10% 부과됩니다.*

![](image/fig 5-1.png)

> *문제2:현빈군은 슈퍼에서 사과를 2개, 귤을 3개 샀습니다. 사과는 1개에 100원, 귤은 1개에 150원입니다. 소비세가 10%일 때 지불 금액을 구하세요.*

![](image/fig 5-3.png)

- 계산 그래프의 처리 순서는 왼쪽에서 오른쪽이며 이를 신경망에 대입하면 **순전파(forward propagation)**라고 함
- 오른쪽에서 왼쪽으로 계산이 처리되는 그 반대 과정을 **역전파(backward propagation)**라고 함
- 역전파는 이후 미분을 계산할 때 중요한 역할을 수행
___

### 2) 국소적 계산

- 국소적 계산이란 '자신과 직접 관계된 작은 범위의 계산'을 의미
- 계산 그래프의 특징은 국소적 계산을 전파하여 최종 결과를 얻는 것에 있음

![](image/fig 5-4.png)

- 위의 그림에서 볼 수 있듯이 3번째 노드에서의 계산은 두 값을 더하는 계산을 수행
- 4,000이라는 값이 어떻게 계산되었는지는 전혀 상관 없이 입력값 2개을 더하기만 하면 문제 해결
- 국소적 계산은 단순하지만, 그 결과를 전달함으로써 전체를 구성하는 복잡한 계산을 해낼 수 있음 (Like assembly line)
___

### 3) 왜 계산그래프로 문제를 푸는가?

- 국소계산으로 **문제를 단순화** 할수 있고, **중간 계산 결과를 보관** 할 수 있는 장점 보유
- 정말 중요한 이유는 **역전파를 통해 '미분'을 효율적으로 계산 가능**


- 문제1에서 사과 가격이 최종 금액에 미치는 영향을 알고 싶다면 **'사과 가격에 대한 지불금액의 미분'**을 구해 설명 가능
- 그림으로 표현하면 아래와 같이 설명 가능
![](image/fig 5-5.png)
- 여기서 역전파는 순전파와 반대방향의 굵은 화살표로 표시하고 국소미분 값을 아래에 표기 (사과 값이 아주 조금 오르면, 최종 금액은 약 2.2배만큼 증가)
- 사과 가격에 대한 미분 외에 '소비세에 대한 지불금액의 미분'이나 '사과 개수에 대한 지불금액의 미분'도 효율적으로 계산 가능 (각 변수의)
