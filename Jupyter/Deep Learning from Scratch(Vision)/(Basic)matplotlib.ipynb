```python
import numpy as np
import matplotlib.pyplot as plt
```

> ### 단순한 그래프 그리기

```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

> ### pyplot의 기능

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y1, label = 'sin')
plt.plot(x, y2, label = 'cos', linestyle = '--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend()
plt.show()
```
