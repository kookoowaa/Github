```python
from lxml.html import parse
from io import StringIO

text = requests.get('http://finance.yahoo.com/q/op?s=AAPL+Options').text
parsed = parse(StringIO(text))

doc = parsed.getroot()
```

```python
from bs4 import BeautifulSoup as bs
import urllib
```

```python
wp = bs(urllib.request.urlopen('http://finance.yahoo.com/q/op?s=AAPL+Options'), 'html.parser')
```

```python
call_data, put_data = wp.find_all('table')
```

```python
columns = [i.get_text() for i in call_data.find_all('th')]
columns
```

```python
call_dict = {}
for i in range(11):
    call_dict[columns[i]] = [i.get_text() for i in call_data.find_all(attrs= {'class' : "data-col{}".format(i)})]
call_table = pd.DataFrame(call_dict)
```

```python
put_dict = {}
for i in range(11):
    put_dict[columns[i]] = [i.get_text() for i in put_data.find_all(attrs= {'class' : "data-col{}".format(i)})]
put_table = pd.DataFrame(put_dict)
```

```python
call_table.sort_values(by = 'Last Trade Date', ascending = False).set_index('Last Trade Date')
```
