# 6장 데이터 로딩, 저장, 파일형식
## HTML, 웹 API와 함께 사용하기

```python
import requests
import pandas as pd
import numpy as np
```

```python
url = 'https://api.github.com/repos/pydata/pandas/milestones/28/labels'

resp = requests.get(url)

resp
```

- Response 객체의 메서드는 파이썬 객체인 사전형태로 변환된 JSON 데이터를 반환

```python
data = resp.json()
data[:5]
```

- 상기 data를 가지고 바로 DataFrame 생성 가능

```python
issue_labels = pd.DataFrame(data)
issue_labels
```

- 평범한 웹 API를 위한 고수준의 인터페이스를 만들어서 DataFrame에 저장 및 분석 수행 가능
