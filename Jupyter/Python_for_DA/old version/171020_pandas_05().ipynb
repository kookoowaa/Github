```python
from pandas import DataFrame, Series
import pandas as pd
import numpy as np
```

```python
frame = DataFrame({'a': range(7), 'b': range(7,0,-1),
                  'c': ['one', 'one', 'one', 'two', 'two', 'two', 'two'],
                  'd': [0, 1, 2, 0, 1, 2, 3]})
frame
```

```python
frame2 = frame[['c', 'd']]
frame2
```

```python
frame3 = frame.set_index(['c','d'])
frame3
```

```python
frame4 = frame.set_index(['c','d'], drop = False)
frame4
```

```python
frame['e'] = list('abcdefg')
frame
```

```python
frame = frame.set_index('e')
frame
```
